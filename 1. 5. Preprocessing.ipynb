```python
import numpy as np
import os
import xmltodict
import pickle
import cv2
import shutil
import matplotlib.pyplot as plt
from random import shuffle
```

```python
# View training images of Akshit after preprocessing (padding on right and bottom to preserve coordinates)
```

```python
path = "./data/training/Akshit/jpg"
jpg_files = os.listdir(path)

for i, jpg in enumerate(jpg_files):
    path_img = os.path.join(path, jpg)
    img = cv2.imread(path_img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    img = cv2.copyMakeBorder(
        img, 
        0, 
        5000-img.shape[0], 
        0, 
        5000-img.shape[1], 
        cv2.BORDER_CONSTANT
    )
    
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    cv2.imwrite(path_img, img)
```

```python
path = "./data/training/Other/jpg"
jpg_files = os.listdir(path)

for i, jpg in enumerate(jpg_files):
    path_img = os.path.join(path, jpg)
    img = cv2.imread(path_img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    img = cv2.resize(img, (5000, 5000))
    
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    cv2.imwrite(path_img, img)
```

```python
# View training images of Akshit with bounding box
```

```python
def rectangle(coords): 
    x = coords["x_c"]
    y = coords["y_c"]
    w = coords["w"]
    h = coords["h"]
    
    x1 = int(x + h)
    x2 = int(x - h)
    y1 = int(y + w)
    y2 = int(y - w)

    return [(x1, y1), (x2, y2)]
```

```python
def draw_rectangle(img, coords):
    coords = rectangle(coords)
    print(coords)
    
    cv2.rectangle(
        img,
        pt1 = coords[0],  
        pt2 = coords[1],
        color = (225, 225, 225),
        thickness = 20
    )
    plt.figure(figsize=(10,10))
    plt.imshow(img)
```

```python
coord = pickle.load(open("./data/training/train_xml_coords.pickle", "rb"))
path = coord[0]["img"]

img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

draw_rectangle(img, coord[0])
```
