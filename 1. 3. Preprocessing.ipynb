```python
import numpy as np
import os
import cv2
import shutil
import matplotlib.pyplot as plt
from random import shuffle, sample
```

```python
# Sixth, copy files in random to folder temp
```

```python
os.mkdir("./temp")
```

```python
images = os.listdir("./random")
for i, img in enumerate(images):
    shutil.copyfile(
        os.path.join("./random", img), 
        os.path.join("./temp", str(i+1) + ".jpg")
    )
```

```python
# View original images of other
```

```python
images = os.listdir("./temp")

nrow, ncol = 4, 4
fig, axs = plt.subplots(nrow, ncol, figsize = (8, 8))

sample_images = sample(images, nrow*ncol)
sample_images = np.array(sample_images)
sample_images = sample_images.reshape((nrow, ncol))

for i in range(nrow):
    for j in range(ncol):
        path = os.path.join("./temp", sample_images[i, j])
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        axs[i, j].imshow(img)
        axs[i, j].set(xticks = [], yticks = [])

fig.tight_layout()
```

```python
# Seventh, partition other for training and validation
```

```python
jpg = os.listdir("./temp")
shuffle(jpg)

x = len(jpg)*0.8
x = int(x)
train_images = jpg[:x]
valid_images = jpg[x:]
```

```python
for img in train_images:
    path = os.path.join("./temp", img)
    shutil.move(path, "./data/training/Other/jpg")
```

```python
for img in valid_images:
    path = os.path.join("./temp", img)
    shutil.move(path, "./data/validation/Other/jpg")
```
