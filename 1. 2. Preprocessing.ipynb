```python
import xmltodict
import cv2
import os
import shutil
import matplotlib.pyplot as plt
from random import shuffle
```

```python
# View original images of Akshit
```

```python
img = cv2.imread("./data/jpg/1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10, 10))
plt.imshow(img)
```

```python
# View bounding boxes around original images of Akshit
```

```python
f = open("./data/xml/1.xml")
data_dict = xmltodict.parse(f.read())
coords = data_dict["annotation"]["object"]["polygon"]["pt"]

temp = dict(coords[0])
x1 = int(temp["x"])
y1 = int(temp["y"])
    
temp = dict(coords[2])
x2 = int(temp["x"])
y2 = int(temp["y"])

cv2.rectangle(
    img,
    pt1 = (x1, y1), 
    pt2 = (x2, y2),
    color = (225, 225, 225),
    thickness = 20
)
plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
# Fourth, partition jpg and xml to obtain the following structure in data:
# 1. training
# 1.1. Akshit
# 1.2. Other
# 2. validation
# 2.1. Akshit
# 2.2. Other
```

```python
os.mkdir("./data/training")
os.mkdir("./data/training/Akshit")
os.mkdir("./data/training/Other")

os.mkdir("./data/training/Akshit/jpg")
os.mkdir("./data/training/Akshit/xml")
os.mkdir("./data/training/Other/jpg")

os.mkdir("./data/validation")
os.mkdir("./data/validation/Akshit")
os.mkdir("./data/validation/Other")

os.mkdir("./data/validation/Akshit/jpg")
os.mkdir("./data/validation/Akshit/xml")
os.mkdir("./data/validation/Other/jpg")
```

```python
jpg = os.listdir("./data/jpg")
shuffle(jpg)

x = len(jpg)*0.8
x = int(x)
train_images = jpg[:x]
valid_images = jpg[x:]
```

```python
for img in train_images:
    path = os.path.join("./data/jpg", img)
    shutil.move(path, "./data/training/Akshit/jpg")
    
    path = os.path.join("./data/xml", img[:-4] + ".xml")
    shutil.move(path, "./data/training/Akshit/xml")
```

```python
for img in valid_images:
    path = os.path.join("./data/jpg", img)
    shutil.move(path, "./data/validation/Akshit/jpg")
    
    path = os.path.join("./data/xml", img[:-4] + ".xml")
    shutil.move(path, "./data/validation/Akshit/xml")
```

```python
# Fifth, remove jpg and xml since they are redundant
```
