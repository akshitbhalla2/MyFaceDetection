```python
import numpy as np
import os
import pickle
import xmltodict
```

```python
# Ninth, obtain the required data from xml and pickle it
```

```python
def GetXMLCoordinates(typ):
    xml_content = []
    path = "./data/" + typ + "/Akshit/xml"
    xml_files = os.listdir(path)
    
    for file in xml_files:
        f = open(path + "/" + file)
        data_dict = xmltodict.parse(f.read())
        xml_content.append((file, data_dict))
        
    xml_coords = []
    for i, (file, xml) in enumerate(xml_content):
        coords = xml["annotation"]["object"]["polygon"]["pt"]
    
        temp = dict(coords[0])
        x1 = int(temp["x"])
        y1 = int(temp["y"])
    
        temp = dict(coords[2])
        x2 = int(temp["x"])
        y2 = int(temp["y"])
    
        x_c  = (x1 + x2)/2
        y_c  = (y1 + y2)/2
    
        h = np.abs(x1 - x2)/2
        w = np.abs(y1 - y2)/2
    
        xml_coords.append({
            "img": os.path.join("./data/" + typ + "/Akshit/jpg", file[:-4] + ".jpg"),
            "x_c": x_c,
            "y_c": y_c,
            "h": h,
            "w": w
        })
        
    for img in os.listdir("./data/" + typ + "/Other/jpg"):
        xml_coords.append({
            "img": os.path.join("./data/" + typ + "/Other/jpg", img),
            "x_c": 0.,
            "y_c": 0.,
            "h": 0.,
            "w": 0.
        })
        
    return xml_coords
```

```python
train_xml_coords = GetXMLCoordinates("training")
train_xml_coords
```

```python
pickle.dump(train_xml_coords, open("./data/training/train_xml_coords.pickle", "wb"))
```

```python
valid_xml_coords = GetXMLCoordinates("validation")
valid_xml_coords
```

```python
pickle.dump(valid_xml_coords, open("./data/validation/valid_xml_coords.pickle", "wb"))
```
