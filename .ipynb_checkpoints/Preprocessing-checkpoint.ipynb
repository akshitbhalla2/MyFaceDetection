```python
import numpy as np
import os
import xmltodict
import pickle
import cv2
```

```python
xml_files = os.listdir("./data/xml")
xml_content = []

for file in xml_files:
    f = open("./data/xml/" + file)
    data_dict = xmltodict.parse(f.read())
    xml_content.append(data_dict)
```

```python
xml_coords = []
for i in range(len(xml_content)):
    coords = xml_content[i]["annotation"]["object"]["polygon"]["pt"]
    
    temp = dict(coords[0])
    x1 = int(temp["x"])
    y1 = int(temp["y"])
    
    temp = dict(coords[2])
    x2 = int(temp["x"])
    y2 = int(temp["y"])
    
    x_c  = (x1 + x2)/2
    y_c  = (y1 + y2)/2
    
    h = np.abs(x1 - x2)/2
    w = np.abs(y1 - y2)/2
    
    xml_coords.append({
        "img": i + 1,
        "x_c": x_c,
        "y_c": y_c,
        "h": h,
        "w": w
    })
```

```python
pickle.dump(xml_coords, open("xml_coords.pickle", "wb"))
```

```python
jpg_files = os.listdir("./data/jpg")

for file in jpg_files:
    path_img = os.path.join("./data/jpg", file)
    img = cv2.imread(path_img)
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    img = cv2.copyMakeBorder(
        img, 
        0, 
        5000-img.shape[0], 
        0, 
        5000-img.shape[1], 
        cv2.BORDER_CONSTANT
    )
    
    path_save = os.path.join("./data/jpg", file)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    cv2.imwrite(path_save, img)
```

```python
# Plotting
import matplotlib.pyplot as plt
```

```python
def rectangle(coords): 
    x = coords["x_c"]
    y = coords["y_c"]
    w = coords["w"]
    h = coords["h"]
    
    x1 = int(x_c + h)
    x2 = int(x_c - h)
    y1 = int(y_c + w)
    y2 = int(y_c - w)
    ################ FIX THIS 
    return [(x1, y1), (x2, y2)]
```

```python
def draw_rectangle(img, coords):
    coords = rectangle(coords)
    print(coords)
    
    cv2.rectangle(
        img,
        pt1 = coords[0],  
        pt2 = coords[1],
        color = (225, 225, 225),
        thickness = 20
    )
    plt.figure(figsize=(10,10))
    plt.imshow(image)
```

```python
image = cv2.imread("./data/jpg/1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

draw_rectangle(image, xml_coords[0])
```
